**f- Write a function "compute_gradient" that returns two arrays/imagse: one containing the input image gradient and another the gradient norm. It should take an optional argument "sigma" that gives the scale of the derivative (i.e. the standard deviation of the Gaussian kernel used to blur the image before taking the derivative). What is the size of the output compared to the size of the input?**

```python
def compute_gradient(I,sigma=0,kernel='sobel',output='grad',norm=False):
    # we first apply a gaussian kernel to the image in order to blur it 
    # (default value 0, so no filtering)
    # since it was not specified that we have to build it from scratch, 
    # I will use the function introduced in question d- 
    I_filtered = ndimage.gaussian_filter(I, sigma, mode='constant')
    
    # we can define a kernel for x and for y differences
    # two kernels: sobel or prewitt. Sobel seems to be better at 
    # preserving diagonal edges, so we will use it by default.
    if kernel=='prewitt':
        k_x = np.array([[1,0,-1],
                        [1,0,-1],
                        [1,0,-1]])
        k_y = np.array([[1,1,1],
                        [0,0,0], 
                        [-1,-1,-1]])
    if kernel=='sobel':
        k_x = np.array([[1,0,-1],
                        [2,0,-2],
                        [1,0,-1]])
        k_y = np.array([[1,2,1],
                        [0,0,0], 
                        [-1,-2,-1]])
    
    # and convolve the image with the previous kernels
    partial_x = ndimage.convolve(I_filtered,k_x) # derivatives according to x
    partial_y = ndimage.convolve(I_filtered,k_y) # derivatives according to y

    # and compute the gradient norm
    grad_norm = np.sqrt(partial_x**2 + partial_y**2)
    
    if norm: # if we want to normalize the norm of the gradient (for more 
             # pertinent thresholding down the line)
        min_val, max_val= np.min(grad_norm), np.max(grad_norm)
        grad_norm = (grad_norm - min_val)/(max_val - min_val)
    
    if output=='grad': # in python, gradient usually represented as an array
        return [partial_x, partial_y], grad_norm
    
    if output=='deriv': # easier to manipulate when displaying
        return partial_x, partial_y, grad_norm
```

```python
fig = plt.figure(figsize = (3*5, 3*5)) # since I am displaying 3 images per line, 
                                       # I need to multiply by 3 the size of the 
                                       # images, to have approximately the same size
for (index,img) in enumerate(compute_gradient(I,sigma=0,output='deriv')):
    title = "shape: " + str(img.shape)
    fig.add_subplot(1,3,index+1).set_title(title)
    plt.imshow(img)
plt.show()
```

The size of the outputs are the same as the size of the input: (512,512) arrays

**b- Write a function "nms(gradient)" which takes as input the gradient and outputs a binary images with value 1 only for pixels that correspond to a maximum in the direction of the gradient.**

```python
def nms(gradient): # gradient needs to be in the 'deriv' format
    grad_norm = gradient[2]
    # we first need to compute the gradient direction (orientation)
    grad_direction = np.arctan2(gradient[0], gradient[1])
    
    def directions(d,row,col):
        if (0 <= d < 180/8) or (15 * 180/8 <= d <= 2 * 180):
            return (row, col - 1),(row, col + 1) # N-S
        elif (180 / 8 <= d < 3 * 180 / 8) or (9 * 180 / 8 <= d < 11 * 180 / 8):
            return (row + 1, col - 1),(row - 1, col + 1) # SW-NE
        elif (3 * 180 / 8 <= d < 5 * 180 / 8) or (11 * 180 / 8 <= d < 13 * 180 / 8):
            return (row - 1, col),(row + 1, col) # E-W
        else:
            return (row - 1, col - 1),(row + 1, col + 1) # SE-NW
    
    row, col = grad_norm.shape
    res = np.zeros(grad_norm.shape)
    
    for r in range(1, row - 1):
        for c in range(1, col - 1):
            d = grad_direction[r,c]
            
            prv = grad_norm[directions(d,r,c)[0]]
            nxt = grad_norm[directions(d,r,c)[1]]
            
            if grad_norm[r, c] >= prv and grad_norm[r, c] >= nxt:
                res[r, c] = 1
    return res
```

**c- Combine 'a' and 'b' to extract edges with a gradient norm larger than a given threshold.**

we just need to threshold the gradient norm before doing nms. I reused the expression for thresholding from a- and used it to modify the value of grad_norm (line 4). I also gave res the gradnorm value, since we have thresholded it before

```python
def nms_threshold(gradient,threshold=0.1): # gradient needs to be in the 'deriv' format
    grad_norm = (gradient[2] > threshold).astype('float') # gradient norm thresholding

    # we first need to compute the gradient direction (orientation)
    grad_direction = np.arctan2(gradient[0], gradient[1]) # no need to change
    
    def directions(d,row,col):
        if (0 <= d < 180 / 8) or (15 * 180 / 8 <= d <= 2 * 180):
            return (row, col - 1),(row, col + 1)
        elif (180 / 8 <= d < 3 * 180 / 8) or (9 * 180 / 8 <= d < 11 * 180 / 8):
            return (row + 1, col - 1),(row - 1, col + 1)
        elif (3 * 180 / 8 <= d < 5 * 180 / 8) or (11 * 180 / 8 <= d < 13 * 180 / 8):
            return (row - 1, col),(row + 1, col)
        else:
            return (row - 1, col - 1),(row + 1, col + 1)
    
    row, col = grad_norm.shape
    res = np.zeros(grad_norm.shape)
    
    for r in range(1, row - 1):
        for c in range(1, col - 1):
            d = grad_direction[r,c]
            
            prv = grad_norm[directions(d,r,c)[0]]
            nxt = grad_norm[directions(d,r,c)[1]]
            
            if grad_norm[r, c] >= prv and grad_norm[r, c] >= nxt:
                res[r, c] = grad_norm[r, c]
    return res
```

**d- Add the hysteresis thresholding to 'c' to implement a function computing the Canny edges.**
    
<span style="color:lightgrey">Here is one way to do the hysteresis thresholding. Keep a list of the edges for which you have to visit neighboors. Initialize the list with the edges corresponding to the most discriminative threshold. For each edge you are sure of (i.e. in the list), you have to check if its neighboors are considered edges using the less discriminative threshold. If they are, add them to the output edges and in the list of edges to visit.

```python
def canny_edges(I, sigma=1, low=0.2, high=0.4,neighbour_range=[-1,2]):
    dim = I.shape
    grad = compute_gradient(I, sigma, output='deriv')    
    
    nms_high = nms_threshold(grad,high)
    nms_low = nms_threshold(grad,low)
    
    to_visit = []
    visited = np.zeros((dim[0],dim[1]))
    
    for (i,j),val in np.ndenumerate(nms_high):
        if val > high:
            to_visit.append((i,j))
            visited[i,j] = 1
    
    def neighbour(x,y,neighbour_range):
        x_valid_dim = range(0,dim[0])
        y_valid_dim = range(0,dim[1])
        neighbour_range = range(neighbour_range[0],neighbour_range[1])
        list_neighbours = []

        # check neighbours around x,y (for neighbour_range = [-1,2], 3x3 view around x,y: 8 points)
        for i in neighbour_range: 
            for j in neighbour_range:
                if (i+x) in x_valid_dim and (j+y) in y_valid_dim and visited[i+x,j+y] == 0:
                    list_neighbours.append((i+x,j+y))
        return list_neighbours
    
    while to_visit:
        x,y = to_visit.pop()
        neighbours = neighbour(x,y,neighbour_range)
        
        while neighbours:
            i,j = neighbours.pop()
            if nms_low[i,j] > 0:
                nms_high[i,j] += nms_low[i,j]
                to_visit.append((i,j))
            visited[i,j] = 1
        
    return nms_high
```
